```python
from OptionPricing import *
```

```python
asset = Asset('AAPL', 40, dividend_yield=0.2)
int_rate = 0.05
vol = 0.25
T = 2
ot = 'call'
# For loop based on the exercise price
for i in range(25, 61):
    american = AmericanOption("AAPL_CALL", asset, exercise_price=i,
                              option_type=ot, maturity_time=T)
    a = american.binomial_value(100, interest_rate=int_rate,
                                volatility=vol, method='equal probability')
    euro = EuropeanOption("AAPL_CALL", asset, exercise_price=i,
                          option_type=ot, maturity_time=T)
    eb = euro.binomial_value(100, interest_rate=int_rate, volatility=vol,
                             method='equal probability')
    em = euro.monte_carlo_value(1000, path_length=100,
                                interest_rate=int_rate, volatility=vol)
    asian = AsianEuropeanOption("AAPL_CALL", asset, exercise_price=i,
                                option_type=ot, maturity_time=T,
                                averaging_method='geometric')
    ae = asian.monte_carlo_value(1000, path_length=100,
                                 interest_rate=int_rate, volatility=vol)
    print(i)
    print('American Option:', a)
    print('Euro Binom Optn:', eb)
    print('Euro Monte Optn:', em)
    print('Asian Euro Optn:', ae)
```
